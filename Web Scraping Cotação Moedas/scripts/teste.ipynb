```python
class cotacao():
    
    def __init__(self):
        self.leitura_base()
        self.raspagem()
        
    def leitura_base(self):
        import pandas as pd
        path = '../database/moedas.csv'
        df = pd.read_csv(path, sep=';', encoding='latin-1')

        lista = []
        for indice, linha in df.iterrows():
            lista.append(linha['moeda'])
    
            
        self.dados_moedas = lista
        
    def raspagem(self):
        from bs4 import BeautifulSoup
        import requests
        import urllib.parse
        
        headers = {
            'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Mobile Safari/537.36'}
           
        i = 0
        while i < len(self.dados_moedas):
            print(f'Moeda: {self.dados_moedas[i]}')
            moeda_formatada = urllib.parse.quote(self.dados_moedas[i])
            link = f'https://www.google.com/search?q=cotacao+{moeda_formatada}'
            print(link)
            
            requisicao = requests.get(link, headers=headers)
            
            if requisicao.status_code == 200:
                print(f'Sucesso na requisição')
            else:
                print(f'Insucesso na requisição {requisicao.status_code}')
                
            site = BeautifulSoup(requisicao.text, "html.parser")
            cotacao_moeda = site.find('span', class_='DFlfde SwHCTb')
            print(f'1 Real = {cotacao_moeda.get_text()} {self.dados_moedas[i]}')
            print(f'-'*30)
                      
            i += 1      
```
